```python
import pandas as pd
```

```python
# Creating a dummy dataframe
data = {
    'Screen': ['Big', 'Big', 'Medium', 'Small', 'Small'],
    'Model': ['iPhone 12', 'Galaxy S20', 'Pixel 5', 'iPhone 12', 'Galaxy S20'],
    'Defect': ['Yes', 'No', 'No', 'No', 'Yes']
}

df = pd.DataFrame(data)
df
```

```python
# Dictionary
 {  'Small':0, 'Medium': 1, 'Big':2  }
```

```python
# Ordinal
df.Screen.replace({'Small':0, 'Medium': 1, 'Big':2}, inplace=True)
```

```python
df
```

df.Defect.replace{'Yes' : 1, 'No' : 0}, inplace=True}

```python
df.Defect.replace({'Yes' : 1, 'No' : 0}, inplace=True)
```

```python
df
```

```python
df.Defect.replace({1:0,0:1},inplace=True)
df
```
